```python
#List all IPs
import re
import urllib.parse
from iteration_utilities import deepflatten
ip_list = []
def getip(domain_url):
    for each in domain_url:
        ip =re.findall( r'[0-9]+(?:\.[0-9]+){3}', each)
        if len(ip):
            ip_list.append(ip)
        else:
            netloc = urllib.parse.urlparse(each)
            ip_list.append(netloc[1])
    
    return ip_list

getip(url)
print(len(url))
```

```python
#Flatten the list
def flat(l:list):
    for _ in l:
        if isinstance(_,list):
            yield from flat(_)
        else:
            yield _
ip_list = [_ for _ in flat(ip_list)]
ip_list
print(len(ip_list))
```

```python
#the method of getting the ASN
asndb = pyasn.pyasn("ipasn_20150113.dat")
ASN_number = []
ASN_URL = []
def getting_ASN(ip):
    for each in ip:
        try:
            ASN = asndb.lookup(each)
            ASN_number.append(ASN[0])
        except ValueError:
            ASN_URL.append(each)
 
```

```python
##############################################
#drop the URL address that cannot resolve the IP, leaving the IP address that can be resolved
import time
#import eventlet# import eventlet to limit the time for execution
#Because if socket. gethostbyName can get the IP address, it only takes less than 0.3 seconds, and more than 0.3 seconds is erroe.
#eventlet.monkey_patch()
#with eventlet.Timeout(0.1,False)
ip_list_URL = []
def URL_clean(url):
    for each in url:
        #with eventlet.Timeout(0.1,False):
        try:
            url_ip = socket.gethostbyname(each)
                #url_ip_ASN = asndb.lookup(each)
            ip_list_URL.append(url_ip)
        except socket.gaierror:
            continue
```
